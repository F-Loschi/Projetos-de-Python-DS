Agora em Json

```python
dados_pacientes=pd.read_json('/content/pacientes.json')
dados_pacientes
```

Esse veio bonitinho, mas o ngc pode vir aninhado

```python
dados_pacientes_2 = pd.read_json('/content/pacientes_2.json')
dados_pacientes_2
```

  Viu, que feio fica, né?\
Vamo ter que normalizar ela:

O 'pd.json_normalize' é uma função, nos parâmetros, primeiro vem o que tu quer normalizar e depois o que você quer normalizar dele

```python
df_normalizado = pd.json_normalize(dados_pacientes_2['Pacientes'])
df_normalizado
```

Lembrando que esse atalho serve para caso a gente queira normalizar uma única coluna, se quisessemos normalizar mais ainda de uma só ver, faríamos da seguinte maneira:

```python
#Importando o módulo JSON
import json
#Lendo o arquivo json usando o módulo Python JSON
with open('pacientes_2.json','r') as f:
    dados = json.loads(f.read())
#Normalizando os dados com os parâmetros record_path e meta
df_normalizado_2 = pd.json_normalize(dados, record_path='Pacientes', meta=['Pesquisa', 'Ano'])
df_normalizado_2
```

No entanto, a função json_normalize() aceita apenas um dicionário ou uma lista de dicionários. Então, se tentarmos usar parâmetros da função json_normalize em um arquivo JSON podem surgir erros. Por isso usamos primeiro a biblioteca JSON do python pra tratar ela

Vamos salvar agora:

```python
df_normalizado_2.to_json('Pacientes_Normalizados.json')
```
