```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:])
dados
```

```python
dados.reset_index(inplace=True, drop=True)
dados.head()
```

```python
dados.info()
```

Vamos usar o "astype" da numpy para converter o tipo do dado

```python
import numpy as np
```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

Precisamos fazer o tratamento da coluna preços, devido aos caracteres que ela possue

```python
dados['preco']
```

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

Agora vamos fazer o tratamentos das colunas de taxa simultaneamente para depois fazer a conversão

```python
dados[['taxa_deposito','taxa_limpeza']]
```

Usamos o método applymap para ler o datamap inteiro (de acordo com o notebook, o método applymap não é mais usado e sim o método map, o que pode gerar problemas em versões futuras)

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].map(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```
